## Segmentation and Clustering Neighbourhood in Toronto

Let,s scrap the webpage <a href="https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"> Neighbourhood in Toronto </a> using <a href ="http://beautiful-soup-4.readthedocs.io/en/latest/" >beautiful-soup-4 </a> and create our dataframe.

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

import json
from pandas.io.json import json_normalize # tranform JSON file into a pandas dataframe
from geopy.geocoders import Nominatim
import matplotlib.cm as cm
import matplotlib.colors as colors
from sklearn.cluster import KMeans
import folium # map rendering library
```

```python
url ='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
page = requests.get(url).text
soup = BeautifulSoup(page,'lxml')
```

```python
table = soup.find('table')
head = soup.find('tbody').tr
row_lst = head.find_all('th')
```

```python
Postal = row_lst[0].text.split("\n")[0]
Borough = row_lst[1].text.split("\n")[0]
Neighbour = row_lst[2].text.split("\n")[0]
```

```python
postal_list = []
borough_list = []
neighbour_list = []

for rows in table.find_all('tr')[1:] :
    columns = rows.find_all('td')
    postal_code = columns[0].text.split("\n")[0]
    borough = columns[1].text.split("\n")[0]
    neighbour = columns[2].text.split("\n")[0]
    if all([postal_code,borough,neighbour]):
        postal_list.append(postal_code)
        borough_list.append(borough)
        neighbour_list.append(neighbour)
```

```python
df = pd.DataFrame({Postal:postal_list,Borough:borough_list,Neighbour:neighbour_list})
df.head()
```

###   Use  geopy library to get the Coordinates of Toronto city.

In order to define an instance of the geocoder, we need to define a user_agent. We will name our agent <em>my_explorer</em>, as shown below.

```python
address = 'Toronto, ON'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto City are {}, {}.'.format(latitude, longitude))
```

 ###  Now, Let's get the top 100 venue that are in Toronto within a radius of 500 meters 

```python
def getnearbyVenues(names,latitude,longitude,radius = 500):
    venues_list=[]
    for name,lat,lng in zip(names,latitude,longitude):
            
            print(name)
            # create url to explore the venues
            url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            100 )
            
            results = requests.get(url).json()["response"]['groups'][0]['items']
            
            venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])
        
    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
toronto_venues = getnearbyVenues(names = toronto_df['Neighborhood'],latitude = toronto_df['Latitude'],longitude =toronto_df['Longitude'])
```

##  Now we are ready to analyse the neighbourhood venues

### Preprocessing

```python
toronto_oh_encoded = pd.get_dummies(toronto_venues['Venue Category'])
toronto_oh_encoded.head()
```

There  already exist a __Neighborhood__ column in *toronto_oh_encoded* dataframe<br>
we can check  for duplicate using below code


```python
lst = toronto_venues['Venue Category'].unique()
print('Neighborhood' in lst)
```

Let,s rename duplicate  __Neighborhood__ column with all lowercase alphabets

```python
toronto_oh_encoded.rename({'Neighborhood' : 'neighbor_hood'},inplace = True,axis = 1)
toronto_oh_encoded.head()
```

```python
toronto_oh_encoded.insert(toronto_oh_encoded.shape[1],'Neighborhood',toronto_venues['Neighborhood'])
toronto_oh_encoded = toronto_oh_encoded.iloc[:,::-1]

toronto_oh_encoded.head()
```

```python
toronto_oh_encoded.shape
```

#### Let's replace each venue with mean of its Frequency

```python
toronto_grouped = toronto_oh_encoded.groupby('Neighborhood').mean().reset_index()
toronto_grouped.head()
```

```python
toronto_grouped.shape
```

####  Now, we will select the top five  venue in  each neighbourhood

```python
temp_arr = []
for i,neighbor in enumerate(toronto_grouped.Neighborhood):
    print(f'-->{neighbor}<--\n')
    temp = toronto_grouped.iloc[i][1:].sort_values(ascending  = False)
    values = list(temp[:10].index)
    temp_arr.append(values)
    print(temp[:5],"\n")
    print("-------------------------------------------------------------------------------------------------------------------------------------------------------\n")
```

#### create a separate DataFrame for the  top 10  common venues

```python
colomns = []
for i in range(10):
    colomns.append(f'{i+1} Most Common Venue')

neighborhood_venues_sorted = pd.DataFrame(temp_arr,columns =colomns)
neighborhood_venues_sorted.insert(0,'Neighborhood',toronto_grouped['Neighborhood'])        
neighborhood_venues_sorted.head()
```

```python
neighborhood_venues_sorted.shape
```

## Perform Clustering on Neighbourhood

```python
kcluster = 4
 # K-means clustering
toronto_grouped_clustering  = toronto_grouped.drop('Neighborhood',1)
kmeans = KMeans(n_clusters = kcluster,random_state = 0).fit(toronto_grouped_clustering)
labels =kmeans.labels_
labels[0:10]
```

```python
neighborhood_venues_sorted.insert(0,'Cluster labels',labels,)
toronto_merged = df.merge(neighborhood_venues_sorted,on = 'Neighborhood')
```

```python
toronto_merged.head()
```

```python
print(neighborhood_venues_sorted.shape)
print(toronto_merged.shape)
```

### Let's visualize the map cluster

```python
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)
kclusters = 4
# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(toronto_merged['Latitude'], toronto_merged['Longitude'], toronto_merged['Neighborhood'], toronto_merged['Cluster labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
